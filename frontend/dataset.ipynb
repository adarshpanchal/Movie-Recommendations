```python
import pandas as pd
import pickle
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv("data/tmdb_5000_movies.csv")
credits=pd.read_csv("data/tmdb_5000_credits.csv")
```

```python
movies.head()
credits.head()
```

```python
print(movies.info())
```

```python
print(credits.info())
```

```python
merged=movies.merge(credits,on="title")
merged.info()
```

```python
print(merged.shape)
```

```python
merged=merged[['movie_id','title','genres','overview','keywords','vote_average','cast','crew']]
merged.info()
```

```python
merged['overview']=merged['overview'].fillna('')
```

```python
merged.info()
```

```python
merged.shape
```

```python
merged['genres']
```

```python
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
merged['genres']=merged['genres'].apply(convert)
merged['keywords']=merged['keywords'].apply(convert)
```

```python
merged['genres'].head()
merged['keywords'].head()
```

```python
def top3_cast(text):
    c=[]
    count=0
    for i in ast.literal_eval(text):
        if count<3:
            c.append(i['name'])
            count+=1
        else:
            break
    return c
```

```python
merged['cast']=merged['cast'].apply(top3_cast)
merged['cast'].head()
```

```python
def director(text):
    d=[]
    for i in ast.literal_eval(text):
        if(i['job']=='Director'):
            d.append(i['name'])
    return d
```

```python
merged['crew']=merged['crew'].apply(director)
merged['crew'].head()
```

```python
def extract(text):
    try:
        return text.split()
    except:
        return []
```

```python
merged['overview']=merged['overview'].apply(extract)
merged['overview'].head()
```

```python
def spaces(text):
    result=[]
    for i in text:
        result.append(i.replace(" ",""))
    return result
```

```python
merged['genres']=merged['genres'].apply(spaces)
merged['keywords']=merged['keywords'].apply(spaces)
merged['cast']=merged['cast'].apply(spaces)
merged['crew']=merged['crew'].apply(spaces)
merged['crew'].head()
```

```python
merged['tags']=merged['genres'] + merged['overview']+merged['keywords']+merged['cast']
```

```python
merged['tags'].head()
```

```python
new_df=merged[['movie_id','title','vote_average','tags','crew']]
```

```python
new_df.rename(columns={'crew':'director'},inplace=True)
```

```python
new_df['vote_average']=new_df['vote_average']/10
```

```python
new_df.head()
```

```python
def list_to_text(text):
    return " ".join(text)
```

```python
new_df['tags']=new_df['tags'].apply(list_to_text)
new_df['tags'].head()
```

```python
new_df['tags'] = new_df['tags'].apply(str.lower)
```

```python
new_df['vote_average'] = new_df['vote_average'].replace(1, 0.5)
```

```python
new_df[new_df['title']=='Stiff Upper Lips']['vote_average'].iloc[0]
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(new_df['vote_average'], bins=20)

plt.xlabel("Vote Average")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Ratings")

plt.show()
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
similar=cosine_similarity(vectors)
```

```python
pickle.dump(new_df,open("movies.pkl","wb"))
pickle.dump(similar,open("similar.pkl",'wb'))
```
